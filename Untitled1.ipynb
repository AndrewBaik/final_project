```python
import requests
import os
import base64
```

```python
API_KEY=os.environ.get('GOOGLE_CLOUD_API_KEY')
print(API_KEY)
# https://vision.googleapis.com/v1/images:annotate?key=YOUR_API_KEY
```

```python
with open("./image_files/IMG_2053.jpg", "rb") as image_file:
    encoded_string = base64.b64encode(image_file.read())
print(encoded_string)
body = {
  "requests":[
    {
      "image":{
        "content":encoded_string
      },
      "features":[
        {
          "type":"TEXT_DETECTION",
          "maxResults":1
        }
      ]
    }
  ]
}
response = requests.post('https://vision.googleapis.com/v1/images:annotate?key={}'.format(API_KEY), data=body)
```
